```python
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
model = YOLO('/Users/briankimanzi/Documents/programmingLanguages/PythonProgramming/best.pt')
```

```python
local_data_path = "/Users/briankimanzi/Documents/programmingLanguages/PythonProgramming/Fire_Detection_Project"
```

```python
print('Validating Training Set')
train_results = model.val(data=local_data_path, split='train')

print('Validating Validation Set')
val_results = model.val(data=local_data_path, split='val')

print('Validating Testing Set')
test_results = model.val(data=local_data_path, split='test')

accuracy = [
    train_results.top1 * 100,
    val_results.top1 * 100,
    test_results.top1 * 100
]
labels = ['Training', 'Validation', 'Testing']
plt.figure(figsize=(18, 7))
bars = plt.bar(labels, accuracy, color=['#4285F4', '#34A853', '#FBBC05'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}%', ha='center', fontweight='bold')
    
plt.ylim(0, 110)
plt.ylabel('Accuracy (%)')
plt.title('Model Performance Across Dataset Splits')
plt.grid(axis='y', linestyle='---', alpha=.7)
plt.show()
```
